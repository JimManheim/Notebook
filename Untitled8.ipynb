```python
import matplotlib.pyplot as plt
import csv
import pandas as pd
from scipy.interpolate import splrep, splev

wavelengthlist=[]

    



df = pandas.read_csv('C:/Users/camsl/Desktop/Version 2 11-29 Emission.csv').transpose()
indexlist=[5,9,13,17,21,25,29,33,37,41,45,49,53,57,61,65,69,73,77,81]
templist=[2,10,16,24,-12,-21,-25,-34,-42,-47,-58, -67, -73,-83,-88, -90,-100,-109,-120,-168]
superlist=[]
for row in df.iterrows():
    superlist.append(list(row)[1:])

count=0    
    


plt.figure(1,figsize=(18,12))
for x in indexlist[:5]:
    bspl = splrep([i for i in range(len(superlist[x][0]))],superlist[x][0],s=15)
    bspl_y = splev([i for i in range(len(superlist[x][0]))],bspl)
    plt.plot(superlist[0][0], bspl_y,label=templist[count], marker=count, markevery=50)
    count += 1 
    
plt.xlabel('Wavelength (nm)', fontsize=24)
plt.ylabel('Normalized Intensity (a.u)', fontsize=24)
plt.title('Emission vs Temperature', fontsize = 24)
plt.legend(loc="best", fontsize=18)
plt.ylim(0,1.2)
plt.xlim(500, 900)
plt.show()




plt.figure(2,figsize=(18,12))
for x in indexlist[5:12]:
    bspl = splrep([i for i in range(len(superlist[x][0]))],superlist[x][0],s=15)
    bspl_y = splev([i for i in range(len(superlist[x][0]))],bspl)
    plt.plot(superlist[0][0], bspl_y, label=templist[count], marker=count-5, markevery=50)
    count += 1

plt.xlabel('Wavelength (nm)', fontsize=24)
plt.ylabel('Normalized Intensity (a.u)', fontsize=24)
plt.title('Emission vs Temperature', fontsize = 24)
plt.legend(loc="best", fontsize=18)
plt.ylim(0,1.2)
plt.xlim(500, 900)
plt.show()




plt.figure(3,figsize=(18,12))
for x in indexlist[12:19]:
    bspl = splrep([i for i in range(len(superlist[x][0]))],superlist[x][0],s=15)
    bspl_y = splev([i for i in range(len(superlist[x][0]))],bspl)
    plt.plot(superlist[0][0], bspl_y, label=templist[count], marker=count-12, markevery=50)
    count += 1

plt.xlabel('Wavelength (nm)', fontsize=24)
plt.ylabel('Normalized Intensity (a.u)', fontsize=24)
plt.title('Emission vs Temperature', fontsize = 24)
plt.legend(loc="best", fontsize=18)
plt.ylim(0,1.2)
plt.xlim(500, 900)
plt.show()
```
